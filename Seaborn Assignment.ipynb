## Q1
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


    Seaborn is a popular data visualization library in Python that builds on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Here are five commonly used plots in Seaborn and their uses:

    1.Scatter Plot:

     Uses: Scatter plots are used to visualize the relationship between two continuous variables. They are helpful in identifying patterns, clusters, or correlations in data.
     
    2.Line Plot:

      Uses: Line plots are suitable for displaying trends and changes in data over time or any other ordered variable. They are useful for visualizing time series data or continuous data with an inherent ordering.

    3.Bar Plot:

     Uses: Bar plots are ideal for comparing categorical variables. They display the distribution or frequency of categories and make it easy to identify the differences or similarities between groups.

    4.Histogram:

    Uses: Histograms are used to represent the distribution of a single numerical variable. They display the frequency or count of data points within predefined bins, helping to understand the shape and spread of the data.

    5.Heatmap:

     Uses: Heatmaps are useful for visualizing matrices or tabular data. They use color gradients to represent the values in the matrix, making it easy to identify patterns, clusters, or relationships between variables.

    These are just a few examples of the plots available in Seaborn. The library offers many other types of plots, such as box plots, violin plots, swarm plots, and more, which cater to specific visualization needs.

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.


    Certainly! Here's an example of loading the "titanic" dataset using the load_dataset function from Seaborn and creating two box plots: one with "pclass" on the x-axis and "age" on the y-axis, and another with "pclass" on the x-axis and "fare" on the y-axis:




```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")


sns.boxplot(data=titanic_data, x='pclass', y='age')

plt.title("Box Plot: pclass vs age")


plt.show()

sns.boxplot(data=titanic_data, x='pclass', y='fare')

plt.title("Box Plot: pclass vs fare")

plt.show()
```

     In this example, we use the load_dataset function to load the "titanic" dataset from the Seaborn library. Then, we use sns.boxplot to create two box plots.

     In the first box plot, we specify the dataset (titanic_data), set "pclass" as the x-axis (x='pclass'), and "age" as the y-axis (y='age').

     In the second box plot, we again use the titanic dataset, set "pclass" as the x-axis, and "fare" as the y-axis (y='fare').

     After creating each plot, we set a title for the plot using plt.title to indicate the variables being compared.

     Finally, we use plt.show() to display each box plot. Remember to import the necessary libraries, such as seaborn and matplotlib.pyplot, before running this code.

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

    Certainly! Here's an example of using the "diamonds" dataset from Seaborn to plot a histogram for the 'price' column, with the 'cut' column used as the hue parameter:

```python
import seaborn as sns
import matplotlib.pyplot as plt


diamonds_data = sns.load_dataset("diamonds")


sns.histplot(data=diamonds_data, x='price', hue='cut')


plt.title("Histogram of Price with Cut")


plt.show()
```

    In this example, we use the load_dataset function to load the "diamonds" dataset from the Seaborn library. Then, we use sns.histplot to create a histogram.

    We specify the dataset (diamonds_data) and the column we want to plot on the x-axis (x='price'). We set the 'cut' column as the hue parameter (hue='cut'), which will result in different colored bars in the histogram based on the cut quality of the diamonds.

    After creating the plot, we set a title for it using plt.title. Finally, we use plt.show() to display the histogram.

    Make sure to import the necessary libraries, such as seaborn and matplotlib.pyplot, before running this code.

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
 
    Certainly! Here's an example of using the "flights" dataset from Seaborn to plot a heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt


flights_data = sns.load_dataset("flights")


flights_matrix = flights_data.pivot("month", "year", "passengers")


sns.heatmap(data=flights_matrix, cmap="YlGnBu")


plt.title("Flights Heatmap")

plt.show()
```

    In this example, we use the load_dataset function to load the "flights" dataset from the Seaborn library. Then, we use the pivot function to transform the data into a matrix format, where the rows represent the months, the columns represent the years, and the values represent the number of passengers.

    After that, we use sns.heatmap to create the heatmap. We pass the flights_matrix as the data parameter and specify the colormap (cmap="YlGnBu") to control the color scheme of the heatmap.

    We set the plot title using plt.title, and finally, we use plt.show() to display the heatmap.

    Make sure to import the necessary libraries, such as seaborn and matplotlib.pyplot, before running this code.
