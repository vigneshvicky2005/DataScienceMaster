Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.


    Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in various formats. It provides a wide range of functionalities for creating high-quality plots, charts, and graphs.

    Matplotlib is used for data visualization and analysis, making it easier to understand and interpret complex data. It allows users to create a wide variety of plots, customize them, and integrate them seamlessly into their Python programs, interactive environments, or web applications.

    The Pyplot module is a sub-library of Matplotlib that provides a simple interface for creating and customizing plots. It is often used for quick and easy plotting tasks.

    Here are five types of plots that can be created using the Pyplot module of Matplotlib:

     1.Line Plot: A line plot is used to display data points connected by straight line segments. It is suitable for visualizing trends and patterns over a continuous variable.

     2.Scatter Plot: A scatter plot represents individual data points as markers on a 2D plane. It is useful for exploring the relationship between two continuous variables and identifying clusters or patterns.

     3.Bar Plot: A bar plot displays categorical data with rectangular bars of varying heights or lengths. It is commonly used to compare and visualize the distribution or frequency of different categories.

     4.Histogram: A histogram is used to represent the distribution of a continuous variable by dividing it into bins and displaying the frequency or count of observations within each bin.

     5.Pie Chart: A pie chart is a circular plot divided into sectors, where each sector represents a proportion or percentage of a whole. It is useful for displaying the composition or relative sizes of different categories in a dataset.

     These are just a few examples, and Matplotlib provides many more types of plots to cater to various visualization needs.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

     A scatter plot is a type of plot that displays the relationship between two continuous variables. Each data point is represented as a marker on a 2D plane, with one variable plotted along the x-axis and the other variable plotted along the y-axis. Scatter plots are useful for visualizing patterns, trends, clusters, or the correlation between variables.

     Here's the code to generate a scatter plot using the provided data:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

    In this code, we import the necessary libraries (matplotlib.pyplot and numpy) and set the random seed to ensure reproducibility. We generate two sets of random data using np.random.normal, with a mean of 3, a standard deviation of 2, and the length of the x array.

    Then, we use plt.scatter(x, y) to create the scatter plot, where x represents the data points along the x-axis and y represents the data points along the y-axis.

    Next, we set the title using plt.title('Scatter Plot'), xlabel using plt.xlabel('X'), and ylabel using plt.ylabel('Y') to add labels to the plot.

    Finally, we use plt.show() to display the scatter plot.

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


    The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It helps in organizing multiple plots or visualizations in a single window, allowing for easy comparison and analysis.

    The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot. The index starts from 1 and increases from left to right and top to bottom.

    Here's the code to create four line plots using the subplot() function with the provided data:

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

plt.tight_layout()  # Improves the spacing between subplots
plt.show()
```

     In this code, we import the necessary libraries (matplotlib.pyplot and numpy). Then, we define four sets of data points for the lines (Line 1 to Line 4) using NumPy arrays.

     We use the subplot() function to create a 2x2 grid of subplot. For each subplot, we specify the row, column, and index (ranging from 1 to 4). Inside each subplot, we plot the respective line using plt.plot() and add a title using plt.title().

    Finally, we use plt.tight_layout() to improve the spacing between subplots and plt.show() to display the figure containing all four line plots.

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

     A bar plot is a type of chart that represents categorical data with rectangular bars. Each bar's length or height corresponds to the magnitude of the data being represented. Bar plots are used to compare and visualize the distribution, frequency, or magnitude of different categories or groups.

    A bar plot is particularly useful in the following scenarios:

    Comparing categories: Bar plots provide a visual comparison between different categories or groups, making it easy to identify the differences in their values.

    Displaying frequencies or counts: Bar plots can represent the frequency or count of different categories, making it straightforward to understand the distribution or occurrence of each category.

    Showing trends over time: Bar plots can be used to visualize changes in categorical data over time, where each bar represents a specific time period or interval.

    Now, let's plot a bar plot and a horizontal bar plot using the provided data

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.title('Company Profits')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.show()
```

    In this code, we import the necessary libraries (matplotlib.pyplot and numpy). We define an array company that represents the names of the companies and another array profit that represents their corresponding profits.

    For the bar plot, we use plt.bar(company, profit) to create the vertical bars, where company represents the x-axis categories and profit represents the y-axis values. We add a title using plt.title(), an x-label using plt.xlabel(), and a y-label using plt.ylabel().

    For the horizontal bar plot, we use plt.barh(company, profit) to create the horizontal bars, where company represents the y-axis categories and profit represents the x-axis values. We add the same title, x-label, and y-label as in the bar plot.

    Finally, we use plt.show() to display each plot separately.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

    A box plot, also known as a box-and-whisker plot, is a statistical visualization tool used to display the distribution of a dataset. It provides a concise summary of the data's central tendency, spread, and skewness.

    A box plot is particularly useful for the following purposes:

    Identifying central tendency: A box plot displays the median (50th percentile) of the dataset, which represents the central tendency. The position of the median within the box provides insights into the data's symmetry or skewness.

    Understanding spread and variability: The box in a box plot represents the interquartile range (IQR), which contains the middle 50% of the data. The length of the box indicates the spread or variability of the dataset.

    Detecting outliers: Box plots include whiskers that extend from the box to represent the data's range, excluding outliers. Outliers, which lie outside the whiskers, are marked individually as data points, helping identify potential anomalies.

   

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```

    In this code, we import the necessary libraries (matplotlib.pyplot and numpy). We generate two sets of random data, box1 and box2, using the np.random.normal function.

    Next, we combine the data into a list called data. Each element in the list represents a different box in the box plot.

    We create the box plot using plt.boxplot(data, labels=['Box 1', 'Box 2']), where data contains the data arrays for each box, and the labels parameter assigns labels to each box.

    We add a title using plt.title(), and we label the y-axis as "Values" using plt.ylabel().

    Finally, we use plt.show() to display the box plot.
