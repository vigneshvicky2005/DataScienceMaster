## Q1
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.


To calculate the Pearson correlation coefficient, you need to have paired observations of the two variables: the amount of time students spend studying for an exam and their corresponding final exam scores. Once you have the data, follow these steps to calculate the correlation coefficient:

Calculate the mean (average) of the time spent studying and the mean of the exam scores.
Calculate the difference between each observation of the time spent studying and its corresponding mean.
Calculate the difference between each observation of the exam scores and its corresponding mean.
Multiply the differences calculated in steps 2 and 3 for each observation and sum them up.
Calculate the square root of the sum of squares of the differences from step 4.
Divide the result from step 4 by the product of the standard deviations of the time spent studying and exam scores.
The resulting value will be the Pearson correlation coefficient, which ranges between -1 and +1. A value of -1 indicates a perfect negative linear relationship, +1 indicates a perfect positive linear relationship, and 0 indicates no linear relationship.

Interpreting the result depends on the value obtained:

If the correlation coefficient is close to +1, it indicates a strong positive linear relationship. This means that as the amount of time spent studying increases, the final exam scores tend to increase as well.
If the correlation coefficient is close to -1, it indicates a strong negative linear relationship. This means that as the amount of time spent studying increases, the final exam scores tend to decrease.
If the correlation coefficient is close to 0, it indicates a weak or no linear relationship between the two variables. In this case, the amount of time spent studying does not have a consistent impact on the final exam scores.
Keep in mind that the Pearson correlation coefficient measures only the linear relationship between the variables and does not capture any potential non-linear relationships that might exist.

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

To calculate Spearman's rank correlation coefficient, you need to rank the observations of both variables and then apply the formula. Here's how you can calculate the Spearman's rank correlation coefficient for the given data on sleep and job satisfaction:

Rank the observations for each variable separately. Assign ranks based on the order of the values, starting with 1 for the lowest value and incrementing by 1 for each subsequent value. If there are ties (i.e., multiple individuals with the same value), assign them the average of the ranks they would have received without the ties.

Calculate the differences between the ranks of the corresponding observations for both variables.

Square the differences calculated in step 2.

Sum up the squared differences.

Use the following formula to calculate the Spearman's rank correlation coefficient:

rho = 1 - (6 * sum of squared differences) / (n^3 - n),

where 'n' is the number of observations.

The resulting value of Spearman's rank correlation coefficient, rho, will range between -1 and +1. A value of -1 indicates a perfect negative monotonic relationship, +1 indicates a perfect positive monotonic relationship, and 0 indicates no monotonic relationship.

Interpreting the result:

If the Spearman's rank correlation coefficient, rho, is close to +1, it indicates a strong positive monotonic relationship. This means that as the amount of sleep individuals get each night increases, their overall job satisfaction tends to increase as well.
If the coefficient is close to -1, it indicates a strong negative monotonic relationship. This means that as the amount of sleep individuals get each night increases, their overall job satisfaction tends to decrease.
If the coefficient is close to 0, it indicates a weak or no monotonic relationship between the two variables. In this case, the amount of sleep individuals get each night does not have a consistent impact on their overall job satisfaction.
Spearman's rank correlation coefficient measures the monotonic relationship between variables, which means it captures any form of consistent increasing or decreasing pattern, not necessarily linear.
